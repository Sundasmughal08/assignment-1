```python

GREEN = '\033[92m'
END = '\033[0m'

print("--- QUIZ 1: SOLUTIONS BY CODE ---")

#  Data type of "3.14"
val1 = "3.14"
print(f"1. Type of '3.14' is: {GREEN}{type(val1)}{END}") 

#  Math operation
val3 = (2 + 3) * 4
print(f"3. (2 + 3) * 4 = {GREEN}{val3}{END}")

# String multiplication
val5 = "Hello" * 3
print(f"5. 'Hello' * 3 = {GREEN}{val5}{END}")

#  Type of 5
val8 = 5
print(f"8. type(5) is: {GREEN}{type(val8)}{END}")

# Result of int(3.9)
val9 = int(3.9)
print(f"9. int(3.9) is: {GREEN}{val9}{END}")

#  Boolean check
val10 = False
print(f"10. Boolean value example: {GREEN}{val10}{END}")

print("\n--- QUIZ 2: SOLUTIONS BY CODE ---")

```

**QUIZ 2(DATA STRUCTURE)**

```python
# Tuple creation
my_tuple = (1, 2, 3)
print(f"1. Type of (1, 2, 3) is: {GREEN}{type(my_tuple)}{END}")

# Mutability check (List vs Tuple)
my_list = [1, 2]
my_list[0] = 10  # List can change 
print(f"2. List after change: {GREEN}{my_list}{END} (Mutable)")
print("   Note: Tuple change nahi ho sakta (Immutable)")
```

```python
GREEN = '\033[92m'
RED = '\033[91m'
END = '\033[0m'

print("--- Data Structures Code Test ---")

# Dictionary access
person = {"name": "Alice", "age": 30}
print(f"3a. person['name'] gives: {GREEN}{person['name']}{END}")
print(f"3c. person.get('name') gives: {GREEN}{person.get('name')}{END}")

# List index
my_list = [10, 20, 30]
print(f"4. my_list[1] is: {GREEN}{my_list[1]}{END}")

#Missing key in dictionary
my_dict = {"a": 1, "b": 2}
print("7. Accessing my_dict['c']...")
try:
    print(my_dict["c"])
except KeyError:
    print(f"   Result: {RED}KeyError occurred!{END}")

# Q10: Tuple immutability
my_tuple = (1, 2, 3)
print("10. Trying to change my_tuple[1] = 10...")
try:
    my_tuple[1] = 10
except TypeError:
    print(f"    Result: {RED}Error! Tuples cannot be changed.{END}")
```

```python
GREEN = '\033[92m'
END = '\033[0m'

print("--- Quiz 3: Logic & Loops Testing ---")

# If-Elif-Else Logic
x = 10
print("1. Output of if-elif block:")
if x > 5:
    print(f"   {GREEN}A{END}")
elif x > 8:
    print("   B")
else:
    print("   C")

# Range loop
print("4. Output of for i in range(3):", end=" ")
print(f"{GREEN}", end="")
for i in range(3):
    print(i, end=" ")
print(f"{END}")

#While loop
print("8. Output of while loop:")
i = 1
while i < 4:
    print(f"   {GREEN}{i}{END}")
    i += 1

#Ternary Expression
a, b = 10, 5
result = a if a > b else b
print(f"9. Ternary result: {GREEN}{result}{END}")
```
