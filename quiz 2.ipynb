````python
GREEN = '\033[92m'
RED = '\033[91m'
END = '\033[0m'

print("--- Functions & Scope Code Test ---")

# Q4: Scope Logic Test
x = 10
def my_func():
    x = 5 # Local variable
    print(f"Inside function (Local x): {GREEN}{x}{END}", end=" | ")

print("Executing Q4 logic:")
my_func()
print(f"Outside function (Global x): {GREEN}{x}{END}")

# Q7: Lambda Test
double = lambda x: x * 2
print(f"\n7. Lambda result for 5: {GREEN}{double(5)}{END}")

# Q5 & Q6: Map and Filter
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums))
evens = list(filter(lambda x: x % 2 == 0, nums))
print(f"5. Map (squares): {GREEN}{squared}{END}")
print(f"6. Filter (evens): {GREEN}{evens}{END}")
````
