**ASS 2 (NUMPY WITH MATH WIZARD)**

```python
import numpy as np
```

```python
# 2. Create Arrays
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([10, 20, 30])

```

```python
# 3. Perform Operations (Demonstrating Broadcasting)
print("\n--- Array Math & Broadcasting ---")
print(f"Addition (arr1 + arr2):\n{arr1 + arr2}")
print(f"Subtraction (arr1 - arr2):\n{arr1 - arr2}")
print(f"Multiplication (arr1 * arr2):\n{arr1 * arr2}")
print(f"Division (arr1 / arr2):\n{arr1 / arr2}")

```

```python


# 5. Universal Functions (ufuncs)
arr3 = np.array([1, 4, 9, 16, 25])
print("\n--- Universal Functions on arr3 ---")
print(f"Square Root: {np.sqrt(arr3)}")
print(f"Exponential: {np.exp(arr3)}")
print(f"Sine: {np.sin(arr3)}")


```

```python
# 6. Aggregations
arr4 = np.array([[5, 12, 3], [8, 2, 7]])
print("\n--- Statistics on arr4 ---")
print(f"Sum of all elements: {np.sum(arr4)}")
print(f"Max in each column (axis=0): {np.max(arr4, axis=0)}")
print(f"Min in each row (axis=1): {np.min(arr4, axis=1)}")
```

```python
# 7. Type Casting
arr1_float = arr1.astype(float)
print("\n--- Type Casting ---")
print(f"Original arr1 dtype: {arr1.dtype}")
print(f"Converted arr1_float dtype: {arr1_float.dtype}")
```

**ASS 3(PANDAS DATAFRAME DETECTION)**

```python
import pandas as pd
import numpy as np
```

```python
data = {

'Name': ['Alice','Bob','Charlie','Diana','Eve','Frank'],

'Age': [25,30,np.nan,35,28,np.nan],

'Score': [85,92,78,np.nan,88,95],

'Grade': ['A','A','B','C','B','A']

}
```

```python
df = pd.DataFrame(data)
```

```python
df.head(3)
```

```python
df.shape
```

```python
df.columns
```

```python
df.dtypes
```

```python
#missing data
df.isnull()
```

```python
Age_mean = df['Age'].mean()
df['Age'] = df['Age'].fillna(Age_mean)
print(Age_mean)
```

```python

df.dropna(subset=['Score'], inplace=True)
print("Dropped rows where Score was missing.")

```

```python
# 7. Operations
print("\n--- 7. Operations ---")
# Unique grades
print("Unique Grades:", df['Grade'].unique())

# Grade counts
print("\nGrade Counts:\n", df['Grade'].value_counts())
```

```python
# Sort by Score descending
print("\nSorted by Score (Descending):\n", df.sort_values(by='Score', ascending=False))
```
