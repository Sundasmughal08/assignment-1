**The Personal Assistant(day1_3 basics)**

```python

name = input("What is your name? ")


birth_year = int(input("What year were you born? "))

current_age = int(input("What is your current age? "))

year_at_100 = (100 - current_age) + 2025

lucky_number = float(input("Enter your lucky number: "))

print("\n--- Personal Assistant Summary ---")
print(f"Hello, {name}!")
print(f"You were born in {birth_year}.")
print(f"You are currently {current_age} years old.")
print(f"You will turn 100 in the year {year_at_100}.")
print(f"Your lucky number is {lucky_number}.")
```

**ASSIGNMENT 2 (DAY4_DATA STRUCTURE)**

```python
# 1. Initial shopping list
shopping_list = ["milk", "bread", "eggs", "apples", "chicken"]


print(f"Initial Shopping List: {shopping_list}")
# milk=0, bread=1, eggs=2
print(f"The third item is: {shopping_list[2]}")

print("Adding 'cheese' and 'coffee'...")
shopping_list.append("cheese")
shopping_list.append("coffee")

#  remove one item
print("Removing 'chicken'...")
shopping_list.remove("chicken")

#  Dictionary for price
item_prices = {
    "milk": 2.50,
    "bread": 1.99,
    "eggs": 3.25,
    "apples": 0.99
}

print(f"Updated Shopping List: {shopping_list}")
print(f"Item Prices: {item_prices}")
print(f"The price of bread is ${item_prices['bread']}")


print("\n--- Final Summary ---")
# join() method change list into string
print(f"Items to buy: {', '.join(shopping_list)}")
print(f"Price lookup: {item_prices}")
```

**ASSIGNMENT 3(DAY5_LOOPS CONDITIONS)**

```python

for num in range(1, 21):
    
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")

    elif num % 3 == 0:
        print("Fizz")

    elif num % 5 == 0:
        print("Buzz")
    
    else:
        print(num)

#  number for countdown 
print("\n--- Countdown Part ---")
user_num = int(input("Enter a number for countdown: "))

if user_num > 10:
    print("Number is too large! Program ending.")
else:
    while user_num >= 1:
        print(f"Countdown: {user_num}")
        user_num -= 1 
    
    print("Blast Off!")
```
