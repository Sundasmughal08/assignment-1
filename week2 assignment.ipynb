**ASS 1(DATA ANALYZER)**

```python
import math 


def get_numbers():
    user_input = input("Enter numbers separated by spaces: ")
   
    return [float(x) for x in user_input.split()]

# Mean (Average) 
def calculate_mean(data):
    return sum(data) / len(data)

def calculate_variance(data, mean):
    
    sum_of_squares = sum((x - mean) ** 2 for x in data)
    return sum_of_squares / len(data)

# Standard Deviation
def calculate_std_dev(variance):
    return math.sqrt(variance)

# 5. Main Program
def main():
    
    numbers = get_numbers()
    
    if not numbers:
        print("Koi number enter nahi kiya gaya!")
        return

  
    avg = calculate_mean(numbers)
    var = calculate_variance(numbers, avg)
    std = calculate_std_dev(var)

   
    print("\n--- Statistical Analysis ---")
    print(f"Data: {numbers}")
    print(f"Mean: {avg:.2f}") # .2f ka matlab hai point ke baad 2 numbers
    print(f"Variance: {var:.2f}")
    print(f"Standard Deviation: {std:.2f}")


if __name__ == "__main__":
    main()
```

**ASS 2(STUDENT CLASS SYSTEM)**

```python
# 1. Base Class: Person
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 2. Method to display info
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

# 3. Derived Class: Student (Inherits Person)
class Student(Person):
    def __init__(self, name, age, student_id, major):
       
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major

    # 5. Method Overriding
    def display_info(self):
        super().display_info() # Parent wala name aur age print karega
        print(f"Student ID: {self.student_id}")
        print(f"Major: {self.major}")

# 6. Derived Class: GraduateStudent (Inherits Student)
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
        # Student class ke attributes set karna
        super().__init__(name, age, student_id, major)
        # 7. Naya attribute
        self.thesis_topic = thesis_topic

    # Override again
    def display_info(self):
        super().display_info() # Student wala sara data print karega
        print(f"Thesis Topic: {self.thesis_topic}")

# 8. Objects, methods call
print("--- Person ---")
p1 = Person("John Doe", 45)
p1.display_info()

print("\n--- Student ---")
s1 = Student("Jane Smith", 20, "S12345", "Computer Science")
s1.display_info()

print("\n--- Graduate Student ---")
gs1 = GraduateStudent("Bob Johnson", 27, "G67890", "Physics", "Quantum Computing")
gs1.display_info()
```

**ASS3(SHAPE CALCULATOR WITH MAGIC METHOD)**

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # 2. __str__ method (Object)
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"

    # 3. Area calculate 
    def area(self):
        return self.width * self.height

    # 4. Overloading + operator
    def __add__(self, other):
      
        new_width = self.width + other.width
        new_height = self.height + other.height
        return Rectangle(new_width, new_height)

    # 5. Overloading == operator
    def __eq__(self, other):
        return self.area() == other.area()

# 6. Objects 
rect1 = Rectangle(4, 5)
rect2 = Rectangle(2, 3)
rect4 = Rectangle(3, 2) # Comparisons check karne ke liye aik aur rectangle


print(f"Rectangle 1: {rect1}")
print(f"Rectangle 2: {rect2}")
print(f"Area of rect1: {rect1.area()}")

# 8
rect3 = rect1 + rect2
print(f"rect3 (rect1 + rect2): {rect3}")

# 9. Comparison 
print(f"Are rect1 and rect2 equal? {rect1 == rect2}")
print(f"Are rect2 and rect4 equal? {rect2 == rect4}") # Inka area 6 hai, toh True aana chahiye
```
